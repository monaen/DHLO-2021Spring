# Visualize the Functions

Programming can not only alleviate miscellaneous labor, in some way, it can also help you to comprehend things through visualization. In this script, we will introduce how to use Python to make a plot!

The first step for working on figures is to **import the visualization packages**. As previously introduced, there are two frequently-used packages that handling the figures, i.e. **Matplotlib** and **Seaborn**. In this script, we will use the Matplotlib package. We will demonstrate some basic usages of this packages through examples. However, both the packages are very powerful and they can do a lot of things beyond what we will learn in this script. You can explore their functions through the scripts left as our howework.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
# Note: the line start with % is called magic command, it allow plots to be 
# rendered directly inside the notebook
```

## 1. Visualize the sin() function

```python
x = np.linspace(0, 10, 1000)
y = np.sin(.5 * np.pi * x)
plt.plot(x, y, linewidth=2, linestyle='--')
plt.grid()
plt.xlim([0, 8]);
```

## 2. Visualize the cos() function

```python
x = np.linspace(0, 10, 1000)
y = np.cos(.5 * np.pi * x)
plt.plot(x, y)
plt.grid()
plt.xlim([0, 8]);
```
